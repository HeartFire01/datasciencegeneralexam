### (1) Drop semua kolom yang tidak diperlukan pada data employee.csv. Lakukan EDA univariat untuk setiap kolom numerik pada employee.csv yang mencakup:
    a. histogram dan boxplot untuk tiap kolom<p>
    b. metrik statistik dasar untuk tiap kolom: mean, std, min, q1, q2, q3, iqr, max<p>
    c. identifikasi nilai upper whisker dan lower whisker dari boxplot tiap kolom<p>
    d. apabila terdapat outlier (<q1-1.5*iqr | >q3+1.5*iqr): hitung count, proportion, dan list dari outlier tiap kolom<p>
    e. identifikasi hal yang menurut anda menarik dari hasil EDA yang Anda dapatkan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('employee.csv')
df = df.dropna(axis=1, how='all')
df.head()

# a. histogram dan boxplot untuk tiap kolom
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()

for col in numeric_columns:
    plt.figure(figsize=(12, 6))
    
    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')
    
    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    
    plt.tight_layout()
    plt.show()

# b. metrik statistik dasar untuk tiap kolom: mean, std, min, q1, q2, q3, iqr, max
basic_stats = df[numeric_columns].describe().T

# Calculate IQR and add it to the dataframe
basic_stats['IQR'] = basic_stats['75%'] - basic_stats['25%']

# Displaying basic statistics
print("Basic Statistics (including IQR):")
print(basic_stats[['mean', 'std', 'min', '25%', '50%', '75%', 'IQR', 'max']])

# c. identifikasi nilai upper whisker dan lower whisker dari boxplot tiap kolom
whiskers = {}
for col in numeric_columns:
    Q1 = basic_stats.at[col, '25%']
    Q3 = basic_stats.at[col, '75%']
    IQR = basic_stats.at[col, 'IQR']
    
    lower_whisker = Q1 - 1.5 * IQR
    upper_whisker = Q3 + 1.5 * IQR
    
    whiskers[col] = {'Lower Whisker': lower_whisker, 'Upper Whisker': upper_whisker}


print("\nUpper and Lower Whiskers:")
for col, values in whiskers.items():
    print(f"{col}: Lower Whisker = {values['Lower Whisker']}, Upper Whisker = {values['Upper Whisker']}")

# d. apabila terdapat outlier (<q1-1.5iqr | >q3+1.5iqr): hitung count, proportion, dan list dari outlier tiap kolom
outliers = {}
for col in numeric_columns:
    Q1 = basic_stats.at[col, '25%']
    Q3 = basic_stats.at[col, '75%']
    IQR = basic_stats.at[col, 'IQR']
    
    lower_whisker = Q1 - 1.5 * IQR
    upper_whisker = Q3 + 1.5 * IQR
    
    outlier_data = df[(df[col] < lower_whisker) | (df[col] > upper_whisker)]
    outlier_count = outlier_data.shape[0]
    outlier_proportion = outlier_count / df.shape[0]
    
    outliers[col] = {
        'Jumlah Outlier': outlier_count,
        'Proporsi Outlier': outlier_proportion,
        'Nilai Outlier': outlier_data[col].tolist()
    }

# menampilkan outliers
print("\nOutliers Information:")
for col, info in outliers.items():
    print(f"{col}:")
    print(f"  Jumlah Outlier = {info['Jumlah Outlier']}")
    print(f"  Proporsi Outliern = {info['Proporsi Outlier']:.2%}")
    print(f"  Nilai Outlier: {info['Nilai Outlier'][:5]}...")  # Displaying first 5 outliers

# e. identifikasi hal yang menurut anda menarik dari hasil EDA yang Anda dapatkan

print("\nObservations:")
```

### (5) Lakukan independen t-test (2-sided) dengan ketentuan:

H0: Tidak ada perbedaan mean 'Total Working Years' antara karyawan yang keluar maupun menetap (gunakan kolom 'Attrition' sebagai acuan)

H1: Terdapat perbedaan mean 'Total Working Years' antara karyawan yang keluar maupun menetap (gunakan kolom 'Attrition' sebagai acuan)

alpha = 5%

Print hasil t-test dan tuliskan kesimpulannya.

```python
import pandas as pd
import scipy.stats as stats

# Load the dataset
df = pd.read_csv('employee.csv')

# Memeriksa kolom yang ada dalam dataset untuk memastikan kolom yang akan dihapus benar-benar ada
print(f'Kolom yang ada pada dataset:\n{df.columns}\n')

# Drop unnecessary columns (Ensure the columns exist before dropping them)
cols_to_drop = ['EmployeeCount', 'Over18', 'StandardHours']
df.drop(columns=[col for col in cols_to_drop if col in df.columns], inplace=True)

# Data Preparation: Separate data for Attrition = 'Yes' and Attrition = 'No'
attrition_yes = df[df['Attrition'] == 'Yes']['TotalWorkingYears']
attrition_no = df[df['Attrition'] == 'No']['TotalWorkingYears']

# Perform independent t-test (2-sided)
t_stat, p_value = stats.ttest_ind(attrition_yes, attrition_no)

# Print the results
print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')

# Conclusion based on p-value
alpha = 0.05
if p_value < alpha:
    print("Reject H0: There is a significant difference in the mean 'Total Working Years' between employees who left and stayed.")
else:
    print("Fail to reject H0: There is no significant difference in the mean 'Total Working Years' between employees who left and stayed.")

print("\nKarena p-value jauh lebih kecil dari alpha '(0.05)', kita menolak hipotesis nol '(H0)'. Dengan demikian, kita dapat menyimpulkan bahwa terdapat perbedaan yang signifikan dalam rata-rata 'Total Working Years' antara karyawan yang keluar dan yang tetap bertahan di perusahaan.")
```

### (6) Lakukan one-way ANOVA dengan ketentuan:

H0: Tidak ada perbedaan mean 'Age' antara karyawan dari 3 departemen yang ada di dataset

H1: Setidaknya terdapat 2 departemen yang mean umur karyawannya berbeda

Print hasil one-way ANOVA dan tuliskan kesimpulannya.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('employee.csv')


print(f'Kolom yang ada pada dataset:\n{df.columns}\n')


columns_to_drop = ['EmployeeCount', 'Over18', 'StandardHours']
columns_existing = [col for col in columns_to_drop if col in df.columns]


df.drop(columns=columns_existing, inplace=True)

departments = df['Department'].unique()  
age_by_department = [df[df['Department'] == dept]['Age'] for dept in departments]

# Perform one-way ANOVA
import scipy.stats as stats
f_stat, p_value = stats.f_oneway(*age_by_department)

# Cetak hasil
print(f'F-statistic: {f_stat}')
print(f'P-value: {p_value}')


alpha = 0.05
if p_value < alpha:
    print("Menolak H0: Tidak ada perbedaan signifika pada mean 'Age' diantara employees dalam 3 departemen.")
    print("Dengan demikian, kita dapat menyimpulkan bahwa setidaknya terdapat 2 departemen yang memiliki rata-rata umur karyawan yang berbeda.")
else:
    print("Gagal mMenolak H0: Tidak ada perbedaan signifika pada mean 'Age' diantara employees dalam 3 departemen.")


print("\nKarena P-value = 0.216 yang lebih besar dari 0.05, kita gagal menolak hipotesis nol (H0), yang berarti tidak ada perbedaan yang signifikan dalam rata-rata umur antara karyawan di tiga departemen yang diuji.")
```

### (8) Persiapkan dataset untuk klasifikasi. Jadikan kolom 'Attrition' sebagai target (y). Drop semua kolom yang dianggap tidak diperlukan

```python
# Sel kedua: Persiapkan dataset untuk klasifikasi

# Kolom yang akan dihapus
columns_to_drop = ['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours', 'Attrition']

# Memeriksa apakah kolom-kolom tersebut ada di dalam dataset
columns_to_drop = [col for col in columns_to_drop if col in df.columns]

# Memilih fitur (X) dan target (y)
X = df.drop(columns=columns_to_drop)

# Target variabel (y) adalah kolom 'Attrition'
y = df['Attrition']

# Menampilkan shape dan beberapa baris pertama data fitur dan target
print(f"Shape of X (features): {X.shape}")
print(f"Shape of y (target): {y.shape}")
print("\nFitur pertama dari X:")
print(X.head())
print("\nTarget 'Attrition' pertama:")
print(y.head())
```

### (9) Lakukan train test split, test:test = 4:1, stratify = y.

```python
# Sel ketiga: Train-Test Split dengan test:test = 4:1 dan stratify = y

from sklearn.model_selection import train_test_split

# Melakukan train-test split dengan perbandingan 4:1 dan stratifikasi berdasarkan target (y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Menampilkan dimensi data train dan test
print(f"Dimensi data training (X_train, y_train): {X_train.shape}, {y_train.shape}")
print(f"Dimensi data testing (X_test, y_test): {X_test.shape}, {y_test.shape}")
```

### (10) Lakukan feature engineering yang dianggap diperlukan pada trainset: imputation, encoding, scaling, selection, dll. Lakukan transformasi serupa pada testset tanpa melakukan fitting kembali.

```python
# Sel keempat: Feature Engineering pada Trainset dan Transformasi Serupa pada Testset

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest, f_classif

# Langkah 1: Imputasi, Encoding, Scaling, Selection pada Train Set

# Menentukan kolom numerik dan kategorikal
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()

# Imputer untuk nilai yang hilang pada kolom numerik
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  # Mengisi missing values dengan mean
    ('scaler', StandardScaler())  # Standarisasi nilai numerik
])

# Encoder untuk kolom kategorikal
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Mengisi missing values dengan modus
    ('onehot', OneHotEncoder(handle_unknown='ignore'))  # OneHotEncoding untuk variabel kategorikal
])

# Menerapkan transformer menggunakan ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

# Fitting pipeline untuk train set dan transformasi train set
X_train_transformed = preprocessor.fit_transform(X_train)

# Feature selection (misalnya menggunakan SelectKBest)
selector = SelectKBest(score_func=f_classif, k='all')  # Pilih semua fitur yang relevan berdasarkan ANOVA F-test
X_train_selected = selector.fit_transform(X_train_transformed, y_train)

# Langkah 2: Transformasi serupa pada Test Set tanpa fitting ulang

# Transformasi pada test set (dengan parameter yang sama tanpa fitting ulang)
X_test_transformed = preprocessor.transform(X_test)

# Seleksi fitur pada test set
X_test_selected = selector.transform(X_test_transformed)

# Menampilkan dimensi setelah transformasi
print(f"Dimensi setelah transformasi Train Set: {X_train_selected.shape}")
print(f"Dimensi setelah transformasi Test Set: {X_test_selected.shape}")
```

### 11) Siapkan 3 estimator: lakukan cross-validation dengan estimator Logistic Regression, Decision Tree Classifier, dan XGBoost Classifier (apabila tidak bisa install xgboost silahkan pilih classifier lain untuk menggantikan) untuk menentukan nilai optimal untuk berbagai hyperparameter masing-masing estimator.

```python
# Sel selanjutnya: Cross-validation dengan 3 estimator untuk hyperparameter tuning

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

# Menentukan parameter yang akan diuji untuk masing-masing estimator
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10],  # Regularization strength
    'solver': ['lbfgs', 'liblinear'],  # Solvers untuk logistic regression
    'max_iter': [100, 200, 300]  # Iterasi maksimal untuk konvergensi
}

param_grid_dt = {
    'max_depth': [5, 10, 15, None],  # Kedalaman pohon
    'min_samples_split': [2, 5, 10],  # Jumlah minimal sampel untuk membagi node
    'min_samples_leaf': [1, 2, 4]  # Jumlah minimal sampel di daun
}

param_grid_rf = {
    'n_estimators': [100, 200],  # Jumlah pohon dalam hutan
    'max_depth': [5, 10, 15, None],  # Kedalaman pohon
    'min_samples_split': [2, 5, 10]  # Jumlah minimal sampel untuk membagi node
}

# Membuat list estimator dengan parameter grid masing-masing
estimators = [
    ('Logistic Regression', LogisticRegression(), param_grid_lr),
    ('Decision Tree', DecisionTreeClassifier(), param_grid_dt),
    ('Random Forest', RandomForestClassifier(), param_grid_rf)
]

# Melakukan GridSearchCV untuk masing-masing estimator dan parameter grid
for name, estimator, param_grid in estimators:
    print(f"Performing Grid Search for {name}...\n")
    
    # Cross-validation dengan GridSearchCV
    grid_search = GridSearchCV(estimator, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train_selected, y_train)
    
    # Menampilkan hasil Grid Search terbaik
    print(f"Parameter terbaik untuk {name}: {grid_search.best_params_}")
    print(f"Nilai cross-validation terbaik untuk {name}: {grid_search.best_score_:.4f}")
    
    # Menggunakan model terbaik untuk memprediksi dan menilai akurasi pada test set
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test_selected)
    
    print(f"Laporan Klasifikasi untuk {name}:")
    print(classification_report(y_test, y_pred))
    print("-" * 50)
```

### (12) Fit ketiga estimator dengan trainset. Print classification report untuk trainset dan testset untuk ketiga estimator

```python
# Sel selanjutnya: Fit Ketiga Estimator dengan Trainset dan Print Classification Report

from sklearn.metrics import classification_report

# Fit dan evaluasi ketiga estimator pada train set dan test set
for name, estimator, _ in estimators:
    print(f"Fitting {name}...\n")
    
    # Fitting model pada train set
    estimator.fit(X_train_selected, y_train)
    
    # Prediksi pada train set dan test set
    y_train_pred = estimator.predict(X_train_selected)
    y_test_pred = estimator.predict(X_test_selected)
    
    # Print classification report untuk train set dan test set
    print(f"Laporan Klasifikasi untuk {name} pada Train Set:")
    print(classification_report(y_train, y_train_pred))
    
    print(f"Laporan Klasifikasi untuk {name} pada Test Set:")
    print(classification_report(y_test, y_test_pred))
    print("-" * 50)
```

### (13) Dari hasil performansi yang Anda dapatkan manakah estimator yang paling baik?

```python
performance = []

for name, estimator, _ in estimators:
    print(f"Evaluating {name}...\n")  
  
    estimator.fit(X_train_selected, y_train)
    y_train_pred = estimator.predict(X_train_selected)
    y_test_pred = estimator.predict(X_test_selected)
    
   
    train_report = classification_report(y_train, y_train_pred, output_dict=True)
    test_report = classification_report(y_test, y_test_pred, output_dict=True)
    
   
    train_f1 = train_report['accuracy']
    test_f1 = test_report['accuracy']
    
   
    performance.append({
        'Estimator': name,
        'Train Accuracy': train_f1,
        'Test Accuracy': test_f1
    })


performance_df = pd.DataFrame(performance)

print("Performance comparison:")
print(performance_df)

best_model = performance_df.loc[performance_df['Test Accuracy'].idxmax()]

print(f"\nEstimator terbaik berdasarkan Test Accuracy: {best_model['Estimator']}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from xgboost import XGBRegressor

# 1. Linear Regression atau Polynomial Regression
linear_model = Pipeline([
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),  # Polynomial Features (degree 2)
    ('regressor', LinearRegression())
])

# Parameter untuk mencari hyperparameter optimal
linear_param_grid = {
    'poly__degree': [1, 2, 3],
    'regressor__fit_intercept': [True, False]
}

# 2. Decision Tree, Random Forest atau XGBoost Regressor
# Model Decision Tree Regressor
tree_model = DecisionTreeRegressor()

# Model Random Forest Regressor
rf_model = RandomForestRegressor()

# Model XGBoost Regressor
xgb_model = XGBRegressor()

# Parameter untuk mencari hyperparameter optimal
tree_param_grid = {
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2, 4]
}

rf_param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5]
}

xgb_param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 6, 9],
    'learning_rate': [0.01, 0.1]
}

# Fungsi untuk mencari hyperparameter terbaik menggunakan GridSearchCV
def get_best_model(model, param_grid, X_train, y_train):
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
    grid_search.fit(X_train, y_train)
    return grid_search.best_estimator_, grid_search.best_params_

# Menentukan model terbaik dan hyperparameter terbaik untuk masing-masing regressor
best_linear_model, best_linear_params = get_best_model(linear_model, linear_param_grid, X_train_transformed, y_train_regression)
best_tree_model, best_tree_params = get_best_model(tree_model, tree_param_grid, X_train_transformed, y_train_regression)
best_rf_model, best_rf_params = get_best_model(rf_model, rf_param_grid, X_train_transformed, y_train_regression)
best_xgb_model, best_xgb_params = get_best_model(xgb_model, xgb_param_grid, X_train_transformed, y_train_regression)

# Menampilkan hasil hyperparameter terbaik
print("Model Linear Hyperparameters Terbaik : ", best_linear_params)
print("Model Decision Tree Hyperparameters Terbaik :", best_tree_params)
print("Model Random Forest Hyperparameters Terbaik:", best_rf_params)
print("Model XGBoost Hyperparameters Terbaik:", best_xgb_params)
```

### (18) Print metrics r2, mse, rmse, mae, dan mape (dalam bentuk dataframe) trainset vs testset untuk kedua regressor. Manakah model yang lebih baik performance-nya?

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np
import pandas as pd

# Fungsi untuk menghitung metrics
def calculate_metrics(y_true, y_pred):
    r2 = r2_score(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100  # Mean Absolute Percentage Error
    return r2, mse, rmse, mae, mape

# Model 1: Linear Model (Polynomial Regression)
best_linear_model.fit(X_train_transformed, y_train_regression)
y_train_pred_linear = best_linear_model.predict(X_train_transformed)
y_test_pred_linear = best_linear_model.predict(X_test_transformed)

# Model 2: Decision Tree Regressor (atau Random Forest / XGBoost)
best_tree_model.fit(X_train_transformed, y_train_regression)
y_train_pred_tree = best_tree_model.predict(X_train_transformed)
y_test_pred_tree = best_tree_model.predict(X_test_transformed)

# Menghitung metrics untuk kedua model pada trainset dan testset
metrics_linear_train = calculate_metrics(y_train_regression, y_train_pred_linear)
metrics_linear_test = calculate_metrics(y_test_regression, y_test_pred_linear)

metrics_tree_train = calculate_metrics(y_train_regression, y_train_pred_tree)
metrics_tree_test = calculate_metrics(y_test_regression, y_test_pred_tree)

# Membuat dataframe untuk menampilkan hasil
metrics_df = pd.DataFrame({
    'Model': ['Linear (Train)', 'Linear (Test)', 'Decision Tree (Train)', 'Decision Tree (Test)'],
    'R²': [metrics_linear_train[0], metrics_linear_test[0], metrics_tree_train[0], metrics_tree_test[0]],
    'MSE': [metrics_linear_train[1], metrics_linear_test[1], metrics_tree_train[1], metrics_tree_test[1]],
    'RMSE': [metrics_linear_train[2], metrics_linear_test[2], metrics_tree_train[2], metrics_tree_test[2]],
    'MAE': [metrics_linear_train[3], metrics_linear_test[3], metrics_tree_train[3], metrics_tree_test[3]],
    'MAPE': [metrics_linear_train[4], metrics_linear_test[4], metrics_tree_train[4], metrics_tree_test[4]]
})

# Menampilkan hasil metrics
print(metrics_df)


best_model = 'Linear' if metrics_linear_test[0] > metrics_tree_test[0] else 'Decision Tree'
print(f"Model dengan performa terbaik berdasarkan R² pada test set: {best_model}")
```

### (19) Lakukan transformasi yang diperlukan pada dataset employee

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import pandas as pd

# Membaca dataset
df = pd.read_csv('employee.csv')

# Menampilkan beberapa baris pertama dari dataset untuk memeriksa struktur data
print(df.head())

# Menentukan kolom numerik dan kategorikal
numeric_features = df.select_dtypes(include=['float64', 'int64']).columns
categorical_features = df.select_dtypes(include=['object']).columns


numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  # Imputasi nilai hilang dengan rata-rata
    ('scaler', StandardScaler())  # Scaling variabel numerik
])


categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # Imputasi nilai hilang dengan modus
    ('encoder', OneHotEncoder(sparse_output=False, drop='first'))  # Encoding kategorikal
])

# Menyusun column transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Membuat pipeline untuk transformasi data
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Transformasi dataset (trainset dan testset dapat diterapkan serupa setelah ini)
df_transformed = pipeline.fit_transform(df)

# Menampilkan hasil dimensi dataset setelah transformasi
print(f"Dimensi dataset setelah transformasi: {df_transformed.shape}")
```

### (21) Dengan menggunakan elbow method dan sillhouette score tentukan nilai k optimal untuk pembuatan model kmeans/kmedoids clustering

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Menentukan range untuk K (misalnya 1 hingga 10)
k_range = range(2, 11)

# Elbow Method: Menghitung WCSS untuk berbagai nilai K
wcss = []
for k in k_range:
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(df_transformed)  # Gunakan data yang sudah di-transformasi
    wcss.append(kmeans.inertia_)

# Menampilkan plot Elbow Method
plt.figure(figsize=(8,6))
plt.plot(k_range, wcss, marker='o', linestyle='--')
plt.title('Elbow Method For Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()

# Silhouette Score: Menghitung Silhouette Score untuk berbagai nilai K
from sklearn.metrics import silhouette_score
sil_scores = []
for k in k_range:
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(df_transformed)
    sil_score = silhouette_score(df_transformed, kmeans.labels_)
    sil_scores.append(sil_score)

# Menampilkan plot Silhouette Score
plt.figure(figsize=(8,6))
plt.plot(k_range, sil_scores, marker='o', linestyle='--')
plt.title('Silhouette Score For Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.show()

# Menentukan nilai K optimal berdasarkan elbow dan silhouette score
optimal_k_elbow = k_range[np.argmin(np.diff(wcss)) + 1]  # Elbow method (titik elbow)
optimal_k_silhouette = k_range[np.argmax(sil_scores)]  # Silhouette Score

print(f"Nilai K optimal berdasarkan Elbow Method: {optimal_k_elbow}")
print(f"Nilai K optimal berdasarkan Silhouette Score: {optimal_k_silhouette}")
```

```python
kmeans = KMeans(n_clusters=optimal_k_silhouette, init='k-means++', max_iter=300, n_init=10, random_state=42)
kmeans.fit(df_transformed)  # Fit model dengan data yang sudah di-transformasi

# Menambahkan kolom 'label' pada dataset original
df['label'] = kmeans.labels_

# Menampilkan beberapa baris pertama untuk melihat hasilnya
print(df.head())
```
